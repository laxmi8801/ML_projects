```python
data = [active,positive,cured,death]
labels = 'Active','Positive','Cured','Death'
explode = (0,0,0,1.5)
colors = ['lightblue','lightsteelblue','lightpink','black']
fig1, ax1 = plt.subplots(figsize=(13, 8))
plt.pie(data, labels=labels, autopct='%1.1f%%', startangle=15, shadow = True, colors=colors, explode=explode)
plt.axis('equal')
plt.show()
```
